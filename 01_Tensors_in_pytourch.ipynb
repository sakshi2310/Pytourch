## Redution Operation

```python
e = torch.randint(size=(2,3),low=0,high=10,dtype=torch.float32)
e
```

```python
# sum
print(torch.sum(e))
# columns sum
print(torch.sum(e,dim=0))
# rows sum
print(torch.sum(e,dim=1))

# mean
print(torch.mean(e))
# columns
print(torch.mean(e,dim=0))
# max and min
print(torch.max(e))
print(torch.min(e))
# standard
print(torch.std(e))
# variance
print(torch.var(e))

# argmax
print(torch.argmax(e)) # position of the largest value
# argmin
print(torch.argmin(e)) # position of the lowert value

## Matrix multiplication
print(torch.matmul(e,e.T))



```

```python
# dot product
vt1  =  torch.tensor([1,2,3])
vt2 = torch.tensor([3,4,5])
print(torch.dot(vt1,vt2))

# transpose
print(torch.transpose(e,0,1))

# determinat
e1 = torch.rand(3,3)
print(torch.det(e1))

# inverse
print(torch.inverse(e1))
```

## Comparsion Operator

```python
i = torch.randint(size=(2,3),low=0,high=10,dtype=torch.float32)
j = torch.randint(size=(2,3),low=0,high=10)
print(i)
print(j)
```

```python
# grater than
print(i>j)

# less then
print(i<j)

# euqal to
print(i==j)

# not equal to
print(i!=j)
```

```python
k = torch.randint(size=(2,3),low=0,high=10)
k
```

```python
# log
print(torch.log(k))

# exp
print(torch.exp(k))

# sqrt
print(torch.sqrt(k))

# sigmoid
print(torch.sigmoid(k))

# relu
print(torch.relu(k))
```
