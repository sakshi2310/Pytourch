```python
import pandas as pd
import torch
import torch.nn as nn
```

```python
df = pd.read_csv("breast-cancer.csv")
```

```python
df.drop(columns=['id'],inplace=True)
```

```python
x = df.drop(columns=['diagnosis'])
y = df['diagnosis']
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,random_state=42,test_size=0.20)
```

```python
scalar = StandardScaler()
x_train = scalar.fit_transform(x_train)
x_test = scalar.transform(x_test)
```

```python
x_train
```

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
x_train_tensor = torch.from_numpy(x_train).float()
x_test_tensor = torch.from_numpy(x_test).float()
y_train_tensor = torch.from_numpy(y_train).float()
y_test_tensor = torch.from_numpy(y_test).float()
```

```python
x_train_tensor.shape
```

## Dataset & DataLoader

```python
from torch.utils.data import DataLoader,Dataset

class CustomeDataset(Dataset):
    def __init__(self,featurs,labels):
        self.featurs = featurs
        self.labels = labels

    def __len__(self):
        return len(self.featurs)

    def __getitem__(self, index):
        return self.featurs[index],self.labels[index]
        
```

```python
train_dataset = CustomeDataset(x_train_tensor,y_train_tensor)
test_dataset = CustomeDataset(x_test_tensor,y_test_tensor)
```

```python
train_loader = DataLoader(train_dataset,batch_size=32,shuffle=True)
test_loader = DataLoader(test_dataset,batch_size=32,shuffle=True)
```

```python
class MysimpleNN(nn.Module):
    def __init__(self,num_featurs):
        super().__init__()
        self.linear = nn.Linear(num_featurs,1)
        self.sigmoid = nn.Sigmoid()

    def forward(self,featurs):
        out = self.linear(featurs)
        out = self.sigmoid(out)
        return out
    
    def loss_function(self,y_pred,y):
        espilon = 1e-7
        y_pred = torch.clamp(y_pred,espilon,1-espilon)

        # calculate loss
        loss = -(y_train_tensor * torch.log(y_pred) + (1-y_train_tensor)*torch.log(1-y_pred)).mean()
        return loss
```

```python
epochs = 25
learning_rate=0.1
```

```python
# Create model 
model = MysimpleNN(x_train_tensor.shape[1])

# optmizer
optimizer = torch.optim.SGD(model.parameters(),lr=learning_rate)

# Loass
loss_function = nn.BCELoss()
```

## Training Pipeline

```python
for epoch in range(epochs):
    for batch_featurs,batch_labels in train_loader:
        # forward pass
        y_pred = model(batch_featurs)
        # loss calculate
        loss = loss_function(y_pred, batch_labels.view(-1, 1))
        # clear gradients
        optimizer.zero_grad()
        # backward pass
        loss.backward()
        # paramerts update
        optimizer.step()
    # print
    print(f'epoch :{epoch+1},Loss:{loss.item()}')
```

## Evulation

```python
model.eval()
accuracy_list = []

with torch.no_grad():
    for batch_featurs,batch_labels in test_loader:
        y_pred = model(batch_featurs)
        y_pred = (y_pred > 0.8).float()

        batch_accuracy = (y_pred.view(-1) == batch_labels).float().mean().item()
        accuracy_list.append(batch_accuracy)
# Calculate allover accuracy
overall_accuracy = sum(accuracy_list) / len(accuracy_list)
print(f'Accuracy : {overall_accuracy}')
```
